# Visualize 

```python
# Import Basemap - used for plotting geographical locations
from mpl_toolkits.basemap import Basemap
# Import matplotlib.pyplot - used for plotting and visualization of data
import matplotlib.pyplot as plt
# Import numpy - used for creating coordinates on this plot
import numpy as np
# Import pandas - used to manage structured data in python
import pandas as pd
```

```python
# Note - visualization was completed on local machine
filepath = 'FILE_PATH'
```

```python
df = pd.read_csv(filepath,header=None,names=['latitude','longitude'],index_col=None,usecols=[0,1])
df.head()
```

```python
# Instantiate a figure
fig = plt.figure(figsize=(12,9))

m = Basemap(projection='mill',
           llcrnrlat = 20,
           urcrnrlat = 55,
           llcrnrlon = -135,
           urcrnrlon = -60,
           resolution = 'c')

m.drawcoastlines()

m.drawparallels(np.arange(20,60,5),labels=[True,False,False,False])
m.drawmeridians(np.arange(-130,-60,10),labels=[0,0,0,1])

# Create lists to be used for plotting coordinates
sites_lat_y = df['latitude'].tolist()
sites_lon_x = df['longitude'].tolist()

# Create scatter plot
m.scatter(sites_lon_x,sites_lat_y,latlon=True, s=1)
# , s=500, c=colors, marker='o', alpha=1, edgecolor='k', linewidth=1, zorder=2

plt.title('P2 Step 1 - Device Visualization (Continental US only)', fontsize=20)
# plt.savefig(outpath+'/s1_device_vis_us')

plt.show()
```
