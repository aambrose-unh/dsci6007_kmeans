# Visualize K-Means

Note: Visualization completed on local machine

```python
# Import Basemap - used for plotting geographical locations
from mpl_toolkits.basemap import Basemap
# Import matplotlib.pyplot - used for plotting and visualization of data
import matplotlib.pyplot as plt
# Import numpy - used for creating coordinates on this plot
import numpy as np
# Import pandas - used to manage structured data in python
import pandas as pd
import os
```

## Synthetic

```python
def load_df(main_dir,dataset,k,dist_func): # dist_func can be 'circle' or 'euclid'
    file = main_dir+'\\'+dataset+'\\clusters_'+'k'+str(k)+'_'+dist_func
    for f in os.listdir(file):
        if 'part' in f:
            file+='\\'+f
    return pd.read_csv(file,header=None,names=['cluster','latitude','longitude'])
```

```python
def load_centers(main_dir,dataset,k,dist_func):
    file = main_dir+'\\'+dataset+'\\centers_'+'k'+str(k)+'_'+dist_func
    for f in os.listdir(file):
        if '.csv' in f:
            file+='\\'+f
    return pd.read_csv(file,header=None,names=['latitude','longitude'])
```

```python
# Visualize 
def vis(df,centers,title, us_only=True):
    # (ensure enough local memory is avialable)
    # df = device_df.toPandas()

    # Instantiate a figure
    fig = plt.figure(figsize=(12,9))

    if us_only==True:
        s=5
        m = Basemap(projection='mill',
                   llcrnrlat = 20,
                   urcrnrlat = 55,
                   llcrnrlon = -135,
                   urcrnrlon = -60,
                   resolution = 'c')

        m.drawcoastlines()

        m.drawparallels(np.arange(20,60,5),labels=[True,False,False,False])
        m.drawmeridians(np.arange(-130,-60,10),labels=[0,0,0,1])
        
    else:
        s=1
        m = Basemap(projection='mill',
                   llcrnrlat = -90,
                   urcrnrlat = 90,
                   llcrnrlon = -180,
                   urcrnrlon = 180,
                   resolution = 'c')

        m.drawcoastlines()

        m.drawparallels(np.arange(-90,90,20),labels=[True,False,False,False])
        m.drawmeridians(np.arange(-180,180,30),labels=[0,0,0,1])

    for i in range(df.cluster.max()+1):
        # Create lists to be used for plotting coordinates
        sites_lat_y = df['latitude'][df.cluster == i].tolist()
        sites_lon_x = df['longitude'][df.cluster == i].tolist()

        # Create scatter plot
        m.scatter(sites_lon_x,sites_lat_y,latlon=True,s=s)
        
    m.scatter(list(centers['longitude']),list(centers['latitude']),latlon=True, marker='*', s=300, c='y', edgecolor='k')

    plt.title(title, fontsize=20)
    # plt.savefig('device_vis')

    plt.show()
```

```python
# main_dir = 's3://final-kmeans/results'
main_dir = 'LOCAL_FILE_PATH'
```

```python
# load df from s3
dataset = 'synthetic'
k=2
dist = 'circ'

vis(load_df(main_dir,dataset,k,dist),
    load_centers(main_dir,dataset,k,dist),
    dataset+' k='+str(k)+' distance function= '+dist)
```
